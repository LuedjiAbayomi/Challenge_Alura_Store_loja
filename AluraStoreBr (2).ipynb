```python
import pandas as pd

lojas = {
    'Loja_1': pd.read_csv(url),
    'Loja_2': pd.read_csv(url2),
    'Loja_3': pd.read_csv(url3),
    'Loja_4': pd.read_csv(url4)
}

faturamentos = {}
for nome_loja, df_loja in lojas.items():
    faturamentos[nome_loja] = df_loja['Preço'].sum()

df_faturamento = pd.DataFrame(list(faturamentos.items()), columns=['Loja', 'Faturamento'])

maior_faturamento = df_faturamento.loc[df_faturamento['Faturamento'].idxmax()]
menor_faturamento = df_faturamento.loc[df_faturamento['Faturamento'].idxmin()]

print("=== Comparação de Faturamento ===")
print(f"🔝 Maior faturamento: {maior_faturamento['Loja']} - R${maior_faturamento['Faturamento']:,.2f}")
print(f"🔻 Menor faturamento: {menor_faturamento['Loja']} - R${menor_faturamento['Faturamento']:,.2f}")

print("\n=== Ranking de Lojas por Faturamento ===")
print(df_faturamento.sort_values('Faturamento', ascending=False))
```

# 2. Vendas por Categoria


```python
def faturamento_por_categoria(df):
    faturamento_por_categoria = df.groupby('Categoria do Produto')['Preço'].sum()
    return faturamento_por_categoria
```

```python
import pandas as pd

lojas = {
    'Loja_1': pd.read_csv(url),
    'Loja_2': pd.read_csv(url2),
    'Loja_3': pd.read_csv(url3),
    'Loja_4': pd.read_csv(url4)
}

produtos_por_categoria = {}

for nome_loja, df_loja in lojas.items():

    contagem = df_loja['Categoria do Produto'].value_counts().reset_index()
    contagem.columns = ['Categoria', 'Quantidade']
    produtos_por_categoria[nome_loja] = contagem

for loja, df_resultado in produtos_por_categoria.items():
    print(f"\n🔵 {loja} - Produtos por Categoria:")
    print(df_resultado.sort_values('Quantidade', ascending=False))
```

```python
resultado_final = pd.concat(
    {loja: df.set_index('Categoria') for loja, df in produtos_por_categoria.items()},
    axis=1
).fillna(0)

print("\n⭐ Resumo Consolidado:")
print(resultado_final)
```

# 3. Média de Avaliação das Lojas

```python
import pandas as pd

lojas = {
    'Loja_1': pd.read_csv(url),
    'Loja_2': pd.read_csv(url2),
    'Loja_3': pd.read_csv(url3),
    'Loja_4': pd.read_csv(url4)
}
avaliacoes = {}

for nome_loja, df_loja in lojas.items():
    media_avaliacao = df_loja['Avaliação da compra'].mean()
    avaliacoes[nome_loja] = media_avaliacao
    print(f"{nome_loja}: {media_avaliacao:.2f}")
```

```python
melhor_loja = max(avaliacoes, key=avaliacoes.get)
pior_loja = min(avaliacoes, key=avaliacoes.get)

print(f"\n🏆 Melhor loja: {melhor_loja} com média de {avaliacoes[melhor_loja]:.2f}")
print(f"⚠️ Pior loja: {pior_loja} com média de {avaliacoes[pior_loja]:.2f}")
```

# 5. Frete Médio por Loja


```python
import pandas as pd

data = {
    'Produto': ['Assistente virtual', 'Mesa de jantar', 'Jogo de tabuleiro', 'Micro-ondas', 'Cadeira de escritório'],
    'Categoria do Produto': ['eletronicos', 'moveis', 'brinquedos', 'eletrodomesticos', 'moveis'],
    'Preço': [219.08, 256.35, 279.51, 1009.99, 446.99],
    'Frete': [9.249790, 11.234305, 21.262681, 54.667344, 26.964689],
    'Data da Compra': ['16/01/2021', '18/05/2022', '15/03/2021', '03/05/2022', '07/11/2020'],
    'Vendedor': ['Pedro Gomes', 'Beatriz Moraes', 'João Souza', 'João Souza', 'Larissa Alves'],
    'Local da compra': ['SP', 'RJ', 'DF', 'RS', 'MG'],
    'Avaliação da compra': [4, 1, 1, 4, 5],
    'Tipo de pagamento': ['cartao_credito', 'cartao_credito', 'cartao_credito', 'boleto', 'boleto'],
    'Quantidade de parcelas': [8, 4, 1, 1, 1],
    'lat': [-22.19, -22.25, -15.83, -30.17, -18.10],
    'lon': [-48.79, -42.66, -47.86, -53.50, -44.38]
}

df = pd.DataFrame(data)

frete_por_loja = df.groupby('Local da compra')['Frete'].agg(['mean', 'count']).reset_index()
frete_por_loja.columns = ['Local da compra', 'Frete Médio', 'Quantidade de Vendas']

print("\n📊 RESUMO DE FRETES POR LOJA")
print(frete_por_loja.sort_values('Frete Médio', ascending=False).to_string(index=False))
print(f"\n🔝 Maior frete médio: {frete_por_loja.loc[frete_por_loja['Frete Médio'].idxmax()]['Local da compra']} - R${frete_por_loja['Frete Médio'].max():.2f}")
print(f"🔻 Menor frete médio: {frete_por_loja.loc[frete_por_loja['Frete Médio'].idxmin()]['Local da compra']} - R${frete_por_loja['Frete Médio'].min():.2f}")
```

***RELATÓRIO DE DESEMPENHO COMERCIAL***


O presente relatório consolida as principais métricas de desempenho das lojas, destacando padrões de vendas, avaliação de clientes, faturamento e custos logísticos.

**Desempenho por Categoria de Produtos**<BR>
A análise das vendas por categoria revela que os móveis lideram em volume, com 480 unidades vendidas, representando 26,2% do total. Este desempenho contrasta significativamente com a categoria de instrumentos musicais, que registrou apenas 170 vendas, ficando 73,8% abaixo do produto mais vendido. Essa disparidade sugere que estratégias comerciais diferenciadas podem ser necessárias para impulsionar as categorias com menor desempenho.


**Avaliação das Lojas**<BR>
Em relação à satisfação dos clientes, as lojas 1 e 4 se destacam positivamente, ambas com nota média de 4,0. Por outro lado, a loja 3 apresenta a pior avaliação 2,0, indicando possíveis deficiências na experiência do cliente que merecem atenção imediata. É relevante notar que, embora a loja 3 tenha problemas de avaliação, ela detém uma participação significativa no faturamento total, o que pode sugerir um trade-off entre volume de vendas e satisfação do cliente.

**Faturamento e Distribuição de Receita**<BR>
A loja 1 lidera em faturamento, com R\$ 1.534.509,12, contribuindo com 26,1% do total. A loja 4, segue com a menor receita de R\$ 1.384.497,75 correspondente a 23,6%. A distribuição percentual do faturamento mostra que a loja 3, mesmo com avaliações baixas, responde por 24,9% da receita total, um indicador de que seu potencial pode ser melhor explorado com ajustes na qualidade do atendimento ou pós-venda.

**Análise de Custos Logísticos**<BR>
Os fretes apresentam variação significativa entre regiões. O estado do Rio Grande do Sul (RS) registra o valor mais alto (R\$ 54,67), enquanto São Paulo (SP) tem o mais baixo (R\$ 9,25). As demais regiões mantêm uma média entre R\$ 10,00 e R\$ 30,00. Esse cenário indica que políticas de frete diferenciadas ou parcerias logísticas podem ser estratégicas para equilibrar custos e competitividade, especialmente no RS.

**Conclusões e Recomendações Estratégicas**<BR>
Os dados apontam para oportunidades claras de melhoria: (1) a loja 3 requer ações urgentes para elevar sua avaliação, como treinamento de equipe ou revisão de processos; (2) a categoria de instrumentos musicais pode se beneficiar de campanhas promocionais ou bundling com produtos mais populares; (3) a otimização de fretes, principalmente no RS, pode reduzir barreiras à conversão de vendas. Além disso, a loja 4, com boa avaliação mas menor faturamento, é candidata a investimentos em marketing para alavancar seu potencial.
